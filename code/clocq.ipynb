```python
import json
import requests
```

```python
def get_neighborhood(kb_item, p=1000, include_labels=True):
    info_url = "https://clocq.mpi-inf.mpg.de/api/neighborhood?item=" + kb_item + "&p=1000"
    response = requests.get(info_url)
    result = response.json()
    return result
```
